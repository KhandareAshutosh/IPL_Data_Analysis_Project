```python
dataset= dataset.drop(['Unnamed: 0','extras','match_id', 'runs_off_bat'],axis = 1)
dataset.info()
```

```python
a1 = dataset['venue'].unique()
a2 = dataset['batting_team'].unique()
a3 = dataset['bowling_team'].unique()
a4 = dataset['striker'].unique()
a5 = dataset['bowler'].unique()
```

```python
def labelEncoding(data):
    dataset = pd.DataFrame(data)
    feature_dict ={}
    #features={}
    for feature in dataset:
        if dataset[feature].dtype==object:
            le = preprocessing.LabelEncoder()
            fs = dataset[feature].unique()
            le.fit(fs)
            dataset[feature] = le.transform(dataset[feature])
            feature_dict[feature] = le
            print(feature)
    return dataset

labelEncoding(dataset)
```

```python
ip_dataset = dataset[['venue','innings', 'batting_team', 'bowling_team', 'striker', 'non_striker', 'bowler']]
b1 = ip_dataset['venue'].unique()
b2 = ip_dataset['batting_team'].unique()
b3 = ip_dataset['bowling_team'].unique()
b4 = ip_dataset['striker'].unique()
b5 = ip_dataset['bowler'].unique()
#ip_dataset.head(2)
```

```python
features={}
for i in range(len(a1)):
    features[a1[i]]=b1[i]
for i in range(len(a2)):
    features[a2[i]]=b2[i]
for i in range(len(a3)):
    features[a3[i]]=b3[i]
for i in range(len(a4)):
    features[a4[i]]=b4[i]
for i in range(len(a5)):
    features[a5[i]]=b5[i]
features
```

```python
ip_dataset.drop('non_striker',axis=1,inplace=True)
```

```python
X= ip_dataset.iloc[:,]
y = dataset.y
```

```python
dataset = dataset.drop('y',axis = 1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 0)
```

```python
X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
lin = LinearRegression()
lin.fit(X,y)
```

```python
from sklearn.ensemble import RandomForestRegressor
lin = RandomForestRegressor(n_estimators=100,max_features=None)
model =lin.fit(X_train,y_train)
```

```python
model
```

```python
y_pred = lin.predict(X_test)
score = lin.score(X_test,y_test)*100
print("Score value:" , score)
# Larger score value indicates a better fit
```

```python
def prediction(d):
    lst = pd.read_csv(d)
    lst['bowlers'] = lst[lst.columns[5:]].apply(lambda x: ','.join(x.dropna().astype(str)),axis=1)
    #lst.drop(['Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9'],axis = 1,inplace=True)
    labeled_data=[]
    for i in lst:
        val = lst[i][0]
        if type(val)==str:
            if val in features.keys():
                labeled_data.append(features[val])
        else:
            labeled_data.append(val)
    bowl = lst['bowlers'][0].split(',')
    for i in bowl:
        labeled_data.append(features[i])
    l1 = labeled_data[:6]
    l2 = labeled_data[:5]
    l2.append(labeled_data[6])
    l3 = labeled_data[:5]
  #  l3.append(labeled_data[7])
  #  l4 = labeled_data[:5]
  #  l4.append(labeled_data[8])
  #  l5 = labeled_data[:5]
  #  l5.append(labeled_data[9])
    l1 = np.array(l1).reshape(1,6)
    y_pred = lin.predict(l1)
    print(y_pred)
```
